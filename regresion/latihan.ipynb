```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
jumlahData = 100
x = np.array([i*0.1+np.random.randn() for i in range(jumlahData)])
y = np.array([i*0.1+np.random.randn() for i in range(jumlahData)])
```

```python
plt.axis([0,10,0,10])
plt.scatter(x,y)
```

```python
def persamaan(x:int,gradian:int)->int:
    return x*gradian
```

```python
m_prediksi = 0
y_prediksi = 0
for i in range(jumlahData):
    yPrediksi = persamaan(x[i],m_prediksi)
    yActual = y[i]
    
    error = yActual - yPrediksi
    deltaM = error/x[i]
    m_prediksi += deltaM
    y_prediksi = yPrediksi
    
```

```python
plt.axis([0,10,0,10])
plt.title(f"M prediksi : {m_prediksi}")
plt.scatter(x,y)
plt.plot(x,persamaan(x,m_prediksi))
```
