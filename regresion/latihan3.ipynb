```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data1 = pd.read_csv('./data/tourism_with_id.csv').head(200)
data2 = pd.read_csv('./data/tourism_rating.csv')
data2['Rating_Binary'] = data2['Place_Ratings'] >= 3

# Mengubah tipe data kolom "Rating_Binary" menjadi integer
data2['Rating_Binary'] = data2['Rating_Binary'].astype(int)
```

```python
def rata2(Place_id):
    dataSorting = data2[data2["Place_Id"]==Place_id].mean()
    return dataSorting['Place_Ratings']
```

```python
x = np.array([i for i in range(1,len(data1))])
y = np.array([rata2(i) for i in range(1,len(data1))])
```

```python
plt.axis([0,210,2,4])
plt.scatter(x,y)
garisX = np.linspace(-10,len(x)+10)
garisY = np.array([2.1 for i in range(len(garisX))])
```

```python
mean_x = np.mean(x)
mean_y = np.mean(y)

numerator = np.sum((x - mean_x) * (y - mean_y))
denominator = np.sum((x - mean_x) ** 2)

b1 = numerator / denominator
b0 = mean_y - (b1 * mean_x)

# Membuat prediksi berdasarkan model regresi
predictions = b0 + b1 * x

# Membuat plot data dan regresi
plt.scatter(x, y, label='Data', color='b')
plt.plot(x, predictions, label='Regresi', color='r')
plt.xlabel('Place ID')
plt.ylabel('Rata-rata Rating')
plt.title('Regresi Linier Rating berdasarkan Place ID (1-200)')
plt.legend()
plt.grid(True)
plt.show()

# Menampilkan koefisien regresi
print(f'Koefisien Regresi (b1): {b1}')
print(f'Intercept (b0): {b0}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data contoh dengan place ID dan rating
data = [{'place_id': i + 1, 'rating': np.random.uniform(1, 5)} for i in range(200)]

# Mengumpulkan rata-rata rating berdasarkan place ID
place_ids = [data_point['place_id'] for data_point in data]
ratings = [data_point['rating'] for data_point in data]

# Menghitung rata-rata rating untuk setiap place ID
unique_place_ids = np.unique(place_ids)
average_ratings = [np.mean([ratings[j] for j in range(len(ratings)) if place_ids[j] == place_id]) for place_id in unique_place_ids]

# Memisahkan data menjadi fitur (X) dan target (Y)
X = np.array(unique_place_ids)
Y = np.array(average_ratings)

# Menghitung koefisien regresi linier
mean_x = np.mean(X)
mean_y = np.mean(Y)

numerator = np.sum((X - mean_x) * (Y - mean_y))
denominator = np.sum((X - mean_x) ** 2)

b1 = numerator / denominator
b0 = mean_y - (b1 * mean_x)

# Membuat prediksi berdasarkan model regresi
predictions = b0 + b1 * X

# Membuat plot data dan regresi
plt.scatter(X, Y, label='Data', color='b')
plt.plot(X, predictions, label='Regresi', color='r')
plt.xlabel('Place ID')
plt.ylabel('Rata-rata Rating')
plt.title('Regresi Linier Rating berdasarkan Place ID (1-200)')
plt.legend()
plt.grid(True)
plt.show()

# Menampilkan koefisien regresi
print(f'Koefisien Regresi (b1): {b1}')
print(f'Intercept (b0): {b0}')
```
